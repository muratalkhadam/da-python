```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)


df = pd.read_csv('Amazon.csv', parse_dates=['Date'], index_col='Date')
df
```

Завдання 1

Побудувати графік зміни об’ємів продаж:
а) загальний;

```python
plt.plot(df['Volume'])
plt.title("Загальна зміна об'ємів продажів з роками")
plt.xlabel('Дата')
plt.ylabel("Об'єм продажів")
plt.show()
```

б) за 2006 рік;

```python
year_2006 = df.loc['2006-01-01':'2006-12-31']['Volume']
year_2006
```

```python
plt.plot(year_2006, label='За 2006')
plt.title("Зміна об'ємів продажів за 2006")
plt.xlabel('Дата')
plt.ylabel("Об'єм продажів")
plt.legend(loc='best')
plt.show()
```

в) за лютий 2020 року;

```python
february_2020 = df.loc['2020-02-01':'2020-02-28']['Volume']
february_2020
```

```python
plt.plot(february_2020, label='За лютий 2020')
plt.title("Зміна об'ємів продажів за лютий 2020")
plt.xlabel('Дата')
plt.ylabel("Об'єм продажів")
plt.legend(loc='best')
plt.show()
```

г) за лютий 2008 – травень 2010;

```python
feb2008_may2010 = df.loc['2008-02-01':'2010-05-31']['Volume']
feb2008_may2010
```

```python
plt.plot(feb2008_may2010, label='За лютий 2008 - травень 2010')
plt.title("Зміна об'ємів продажів за лютий 2008 - травень 2010")
plt.xlabel('Дата')
plt.ylabel("Об'єм продажів")
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.show()
```

д) за 2016 та 2018 на одному графіку (паралельно).

```python
df_2016 = df.loc['2016-01-01':'2016-12-31']['Volume']
df_2018 = df.loc['2018-01-01':'2018-12-31']['Volume']
df_2018.index = df_2018.index.map(lambda x: x.replace(year=2016))
```

```python
plt.plot(df_2016, label='2016')
plt.plot(df_2018, label='2018')
plt.title("Зміна об'ємів продажів у 2016 та 2018 роках")
plt.xlabel('Дата')
plt.ylabel("Об'єм продажів")
plt.legend()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m-%d'))
plt.show()
```

Завдання 2

Знайти середні значення ціни на час відкриття біржі:
а) за 2012 рік;

```python
open_2012 = df.loc['2012-01-01':'2012-12-31']['Open']
open_2012
```

```python
print('Середня вартість ціни при відкритті біржі станом на 2012 рік:', round(open_2012.mean(), 2))
```

б) за кожний рік;

```python
avg_per_year = df['Open'].resample('Y').mean()
avg_per_year
```

в) за кожний тиждень четвертого кварталу 2016 року.

```python
avg_week_q4_2016 = df.loc['2016-10-01':'2016-12-31']['Open'].resample('W').mean()
avg_week_q4_2016
```

г) Розрахувати і зобразити зміни ціни на час відкриття біржі у відсотках за кожний день впродовж осені 2011 року.

```python
avg_autumn_2011 = df.loc['2011-09-01':'2011-11-30']['Open']
percentage_change = (avg_autumn_2011 / avg_autumn_2011.shift(1) - 1) * 100
percentage_change
```

```python
plt.plot(percentage_change.index, percentage_change, linestyle='--', color='b')
plt.title('Зміна курсу у відсотковому відношенні за кожний день впродовж осені 2011 року.')
plt.xlabel('Дата')
plt.ylabel('Зміни середніх значення ціни на час відкриття біржі у відсотках')
plt.grid(True)
plt.show()
```

д) Знайти та зобразити графічно ковзне середнє ціни на час відкриття біржі за 2019 рік з вікном в місяць.

```python
avg_open_prices_2019 = df.loc['2019-01-01':'2019-12-31']['Open']
avg_open_prices_2019
```

```python
month_rolling_mean_2019 = avg_open_prices_2019.rolling(window=30).mean()

plt.plot(avg_open_prices_2019, label='Середня ціна на час відкриття за день')
plt.plot(month_rolling_mean_2019, label='Ковзне середнє (місячне вікно)')
plt.title('Ковзне середнє за середньою ціною на час відкриття біржі в 2019 рік з вікном в місяць')
plt.xlabel('Дата')
plt.ylabel('Ціна')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
